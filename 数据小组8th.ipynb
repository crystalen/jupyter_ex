```python
df=pd.read_csv('houses.csv',encoding='gbk')
df
```

```python
price_sum=df['price'].sum()/df['size'].sum()
price_sum
```

```python
data=df.groupby(['huxing']).agg({'size':['count','mean']})
data.sort_values(by=[('size', 'count')],ascending=False)
```

```python
df['xiaoqu_name'].value_counts()
```

```python
data1=df.groupby(df['xiaoqu_name']).agg({'price':'sum','size':'sum'})
data1['avg_price']=data1['price']/data1['size']
data1.sort_values(by=['avg_price'],ascending=False)
```

```python
data1['total_price']=data1['price']*data1['size']
data1['total_price'].sort_values(ascending=False).head(5)
```
