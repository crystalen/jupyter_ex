```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

```python
dic1=[{'id':'1001','date':'2019-01-01','goods':'A','sales':'3000'},
       {'id':'1001','date':'2019-01-01','goods':'B','sales':'2000'},
       {'id':'1002','date':'2019-01-01','goods':'A','sales':'3000'},
       {'id':'1003','date':'2019-01-01','goods':'B','sales':'2000'},
       {'id':'1001','date':'2019-01-02','goods':'A','sales':'3000'},
       {'id':'1002','date':'2019-01-02','goods':'A','sales':'3000'},
       {'id':'1002','date':'2019-01-02','goods':'B','sales':'2000'},
       {'id':'1003','date':'2019-01-02','goods':'A','sales':'3000'},
       {'id':'1001','date':'2019-01-03','goods':'A','sales':'3000'},
       {'id':'1002','date':'2019-01-03','goods':'B','sales':'2000'},
       {'id':'1001','date':'2019-01-04','goods':'A','sales':'3000'},
       {'id':'1002','date':'2019-01-04','goods':'A','sales':'3000'}]
df=pd.DataFrame(data=dic1)
df
```

```python
df.info()
import datetime
df['date']=pd.to_datetime(df['date'])
df['sales']=df['sales'].astype(int)
df
```

```python
pd.pivot_table(df,index=['id','goods'],columns='date',values='sales',aggfunc=np.sum,fill_value=0)
```

```python
pd.pivot_table(df,index=['id','goods'],columns='date',values='sales',aggfunc=np.sum,fill_value=0,margins=True)
```

```python
dic2=[{'sex':'man','age':'15','smoke':'True','height':'168'},
         {'sex':'man','age':'23','smoke':'False','height':'179'},
         {'sex':'women','age':'25','smoke':'False','height':'181'},
         {'sex':'women','age':'17','smoke':'True','height':'166'},
         {'sex':'man','age':'35','smoke':'True','height':'173'},
         {'sex':'women','age':'57','smoke':'False','height':'178'},
         {'sex':'man','age':'24','smoke':'False','height':'188'},
         {'sex':'women','age':'31','smoke':'True','height':'190'},
         {'sex':'women','age':'22','smoke':'False','height':'160'}]
df2=DataFrame(data=dic2)
df2
```

```python
df2[['age','height']]=df2[['age','height']].astype(int)
df2.info()
```

```python
df2.groupby('sex').mean()
```

```python
pd.crosstab(df2.sex,df2.smoke,margins=True)
```

```python
df2['agegroup']=df2['age'].apply(lambda x: '25以下' if x<25 else ('25-35' if x<35 else ('35-45' if x<45 else '45以上')))
df2
```

```python
pd.crosstab(df2.agegroup,df2.smoke,df2.smoke,margins=True)
```
